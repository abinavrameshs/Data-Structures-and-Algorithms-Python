```python
class Node (object) : 
    def __init__(self, value) -> None:
        self.value = value 
        self.next = None 
        self.prev = None 


class DoublyLinkedList(object) : 
    def __init__(self, value) -> None:
        new_node = Node(value)
        self.length = 1 
        self.head = new_node 
        self.tail = new_node 

    def print_list(self) : 
        temp = self.head 
        while temp is not None : 
            print(temp.value)
            temp = temp.next 

    def append(self, value): 
        new_node = Node(value=value )
        # corner case : When there are no items in the list
        if self.length ==0 : 
            self.head = new_node 
            self.tail = new_node 
            
        else : 
            new_node.prev = self.tail 
            self.tail.next = new_node 
            self.tail = new_node 
        self.length+=1
        return True 

    def pop(self) : 
        # if there are no elements in the DLL 
        if self.head is None : 
            return_node = None
        elif self.length == 1 : 
            return_node = self.tail  
            self.head = None 
            self.tail = None
            self.length-=1
        # if there are elements in DLL
        else : 
            return_node = self.tail 
            self.tail = self.tail.prev
            self.tail.next = None 
            return_node.prev = None 
            self.length-=1
        return return_node 

    def prepend(self, value) : 
        new_node = Node(value)
        # No items in the list
        if self.head is None : 
            self.head = new_node 
            self.tail = new_node 
        # multiple elements in the list
        else: 
            new_node.next = self.head 
            self.head.prev = new_node 
            self.head = new_node 
        self.length+=1
        return True

    def pop_first(self): 
        # if there are 0 elements 
        if self.head is None : 
            return_node = None 
        # if there is just one element 
        elif self.length == 1 : 
            return_node = self.head 
            self.head = None 
            self.tail = None 
            self.length-=1
        # if there are multiple elements
        else : 
            return_node = self.head 
            next_node =self.head.next
            next_node.prev = None 
            self.head = next_node 
            return_node.next = None 
            self.length-=1 
        return return_node 
        

    def get(self,index) : 

        # if index is out of bounds
        
        if index <0 or index>=self.length :
            print("Invalid index") 
            return_node = None
        # if index is within bounds
        else : 
            if index<self.length/2 : 
                temp = self.head
                for _ in range(index):
                    temp = temp.next 
                return_node = temp 
            else : 
                temp = self.tail 
                for _ in range(self.length-1 , index, -1): 
                    temp = temp.prev 
                return_node = temp 
        
        return return_node 
        

    def set_value(self, index, value): 

        # Get the index 
        required_node = self.get(index)

        # CHeck if the node is a valid Node or None
        if required_node : 
            required_node.value = value 
            return True 
        return False

    
    def insert(self, index, value) : 

        
        if index == 0 : 
            self.length+=1
            return self.prepend(value)
        elif index == self.length : 
            self.length+=1
            return self.append(value)
        else : 
            # Get the index 
            required_node = Node(value)
            
            prev_node = self.get(index -1)
            next_node = prev_node.next 

            prev_node.next = required_node
            required_node.prev = prev_node 

            required_node.next = next_node
            next_node.prev = required_node
            
            self.length+=1
            return True 

    def remove(self, index): 
        if index == 0 : 
            return self.pop_first()
        elif index == self.length - 1 : 
            return self.pop()
        else : 
            required_node = self.get(index)
            next_node = required_node.next 
            prev_node = required_node.prev 

            prev_node.next = next_node
            next_node.prev = prev_node 

            required_node.next = None 
            required_node.prev = None 

            self.length-=1
            
            return required_node

            






```
